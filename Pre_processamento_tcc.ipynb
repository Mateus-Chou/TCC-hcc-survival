```python
from scipy.stats import chi2_contingency
def cat_x_cat(col):
    try:
        print(str(df[col].isnull().sum()) + " nulos")
        print()
        print(df[col].value_counts())
        combs = [(1,1),(1,0), (0,1), (0,0)]
        dic = {alvo:[], col:[], 'proporcao':[]}
        for i in combs:
            dic[col].append(i[0])
            dic[alvo].append(i[1])
            df_aux = df[[alvo,col]].value_counts().reset_index()

            df_aux.columns = [df_aux.columns[0], df_aux.columns[1], 'count']
            total_alvo = df[~df[col].isnull()][alvo].value_counts()[i[1]]
            col_do_alvo = df_aux[(df_aux[alvo] == i[1]) & (df_aux[col] == i[0])]['count'].values[0]
            dic['proporcao'].append(col_do_alvo/total_alvo)

        proporcoes = pd.DataFrame(dic)
        # Plotar o grafico de barras agrupadas
        fig = px.bar(
            proporcoes,
            x=alvo,
            y='proporcao',
            color=col,
            title='Proporção de '+ col +' por morte',
            barmode='group'
        )

        fig.show()

        matriz_contingencia = pd.crosstab(df[col], df[alvo])

        fig = px.imshow(
            matriz_contingencia.values,
            labels=dict(x=alvo, y=col),
            x=matriz_contingencia.columns,
            y=matriz_contingencia.index,
            color_continuous_scale='purp',
            title="Matriz de Contingencia"
        )
        # Adicionar anotações de valores numéricos
        for i in range(len(matriz_contingencia.index)):
            for j in range(len(matriz_contingencia.columns)):
                fig.add_annotation(
                    x=matriz_contingencia.columns[j],
                    y=matriz_contingencia.index[i],
                    text=str(matriz_contingencia.values[i, j]),
                    showarrow=False
                )

        fig.show()
        print(proporcoes)


        # Realizar o teste Qui-Quadrado
        chi2, p, dof, expected = chi2_contingency(matriz_contingencia)
        print('\nResultado do Teste Qui-Quadrado:')
        print(f'Valor p: {p}')
        print('Tabela de Frequências Esperadas:')
        print(expected)
    except:
        print("sem variedade de valores possiveis")
    #Valor P abaixo de 0,05 mostrando uma significancia grande!
```
